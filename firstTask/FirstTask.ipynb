#把CSV檔讀入

```python
import pandas as pd

df = pd.read_csv('FristTest.csv')

df.head()
```

```python
temp = df[['Complex2Name','FValue']]
temp
```

```python
Type = temp.groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
Type['year'] = 2018
Type
```

 #讀入2016~2018年的預算資料

```python
df2016 = pd.read_csv('2016.csv')
df2017 = pd.read_csv('2017.csv')
```

```python
new = df2016[['Complex2Name','FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
new['year'] = 2016
```

```python
new
```

```python
frames = [Type,new]
result = pd.concat(frames)
```

```python
result
```

```python
second = df2017[['Complex2Name','FValue']].groupby(['Complex2Name'])['FValue'].sum().reset_index()
```

```python
second['year'] =2017
```

```python
second
```

```python
frames = [Type,new,second]
result = pd.concat(frames)
```
